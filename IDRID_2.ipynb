```python
from PIL import Image
import numpy as np
import os

# Define the target image size
target_size = (224, 224)  # Adjust as needed based on your model's input size

# Define the path to the folder containing the images
images_folder = '/content/drive/MyDrive/Arjun/B. Disease Grading/1. Original Images/a. Training Set/'

# Define the path to the folder where preprocessed images will be saved
output_folder = '/content/drive/MyDrive/Arjun/B. Disease Grading/Preprocessed Images/'

# Function to preprocess and save images
def preprocess_and_save_images(images_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    for filename in os.listdir(images_folder):
        if filename.endswith(".jpg"):
            # Construct the full path to the image file
            image_path = os.path.join(images_folder, filename)
            # Preprocess the image
            img = preprocess_image(image_path)
            # Save the preprocessed image
            save_path = os.path.join(output_folder, filename)
            save_image(img, save_path)

# Function to preprocess and normalize images
def preprocess_image(image_path):
    # Open the image file
    img = Image.open(image_path)
    # Resize the image while maintaining the aspect ratio
    img = img.resize(target_size, Image.ANTIALIAS)
    # Convert the image to numpy array
    img_array = np.array(img)
    # Normalize the pixel values to range [0, 1]
    img_array = img_array.astype('float32') / 255.0
    return img_array

# Function to save the preprocessed image
def save_image(img_array, save_path):
    img = Image.fromarray((img_array * 255).astype(np.uint8))
    img.save(save_path)

# Preprocess and save images
preprocess_and_save_images(images_folder, output_folder)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Define input shape
input_shape = (224, 224, 3)  # Adjust according to your image dimensions

# Define the VGG16 model
def create_vgg16_model(input_shape):
    model = models.Sequential()

    # Block 1
    model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=input_shape))
    model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

    # Block 2
    model.add(layers.Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(layers.Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

    # Block 3
    model.add(layers.Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(layers.Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(layers.Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

    # Block 4
    model.add(layers.Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(layers.Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(layers.Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

    # Block 5
    model.add(layers.Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(layers.Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(layers.Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2, 2), strides=(2, 2)))

    # Flatten the output and add dense layers
    model.add(layers.Flatten())
    model.add(layers.Dense(4096, activation='relu'))
    model.add(layers.Dense(4096, activation='relu'))
    model.add(layers.Dense(5, activation='softmax'))  # Assuming 5 classes for retinopathy grades

    return model

# Load training, validation, and test data
def load_data(train_images_folder, train_labels_file, val_images_folder, val_labels_file, test_images_folder, test_labels_file):
    # Load training data
    train_images = np.array([np.array(Image.open(os.path.join(train_images_folder, filename))) for filename in os.listdir(train_images_folder) if filename.endswith(".jpg")])
    train_labels_df = pd.read_csv(train_labels_file)
    train_labels = train_labels_df['Retinopathy grade'].values  # Assuming 'Retinopathy grade' is the label column
    # Load validation data
    val_images = np.array([np.array(Image.open(os.path.join(val_images_folder, filename))) for filename in os.listdir(val_images_folder) if filename.endswith(".jpg")])
    val_labels_df = pd.read_csv(val_labels_file)
    val_labels = val_labels_df['Retinopathy grade'].values  # Assuming 'Retinopathy grade' is the label column
    # Load test data
    test_images = np.array([np.array(Image.open(os.path.join(test_images_folder, filename))) for filename in os.listdir(test_images_folder) if filename.endswith(".jpg")])
    test_labels_df = pd.read_csv(test_labels_file)
    test_labels = test_labels_df['Retinopathy grade'].values  # Assuming 'Retinopathy grade' is the label column
    return train_images, train_labels, val_images, val_labels, test_images, test_labels

# Create the VGG16 model
vgg16_model = create_vgg16_model(input_shape)

# Compile the model
vgg16_model.compile(optimizer=Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Display model summary
vgg16_model.summary()

# Load data
train_images_folder = '/content/drive/MyDrive/Arjun/B. Disease Grading/1. Original Images/a. Training Set/'
train_labels_file = '/content/drive/MyDrive/Arjun/B. Disease Grading/2. Groundtruths/a. IDRiD_Disease Grading_Training Labels.csv'
val_images_folder = '/content/drive/MyDrive/Arjun/B. Disease Grading/1. Original Images/b. Testing Set/'
val_labels_file = '/content/drive/MyDrive/Arjun/B. Disease Grading/2. Groundtruths/b. IDRiD_Disease Grading_Testing Labels.csv'
test_images_folder = '/content/drive/MyDrive/Arjun/B. Disease Grading/1. Original Images/b. Testing Set/'
test_labels_file = '/content/drive/MyDrive/Arjun/B. Disease Grading/2. Groundtruths/b. IDRiD_Disease Grading_Testing Labels.csv'

train_images, train_labels, val_images, val_labels, test_images, test_labels = load_data(train_images_folder, train_labels_file, val_images_folder, val_labels_file, test_images_folder, test_labels_file)

# Train the model
history = vgg16_model.fit(train_images, train_labels, epochs=10, batch_size=32, validation_data=(val_images, val_labels))

# Evaluate the model on test data
test_loss, test_accuracy = vgg16_model.evaluate(test_images, test_labels)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)

# Plot training history
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Predict classes for test data
predictions = vgg16_model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)

# Display classification report
print("Classification Report:")
print(classification_report(test_labels, predicted_classes))

# Display confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(test_labels, predicted_classes))
```
